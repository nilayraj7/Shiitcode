```python
#reverse an integer 
def rev_no(a):
    sign = 0
    if a*(-1)<=0:
        sign = 1
    else:
        sign = -1
#upper half of the code preserves the negative sign and case 0 check 
    a = abs(a)
    s=0
    while a!=0:
        r = a%10
        a = int(a/10)
        s = s*10+r
    return (s*sign)
print(rev_no(1234))
print(rev_no(-245))
print(rev_no(0))
```

```python
def min_max(arr):
   # mi,ma = 0,0 #the problem with this approach can it will start with 0 we have to start with first element of arr
    mi,ma = arr[0],arr[0]
    for x in arr:#for x in arr loops through elements
        if x>ma:
            ma = x 
        if x<mi:
            mi = x 
    return mi, ma
        
min_max([-11,22,23,4,12])
```

```python
#valid pallindrom 
def pall(st):
    t = ''
    for x in range(len(st)-1,-1,-1):
        t = t+st[x]
    if t == st:
        return True 
    else:
        return False
    
print(pall('NILAY'))
print(pall('BAXAB'))
```

```python
#prime check 
def prime(a):
    c=0
    for x in range(1,a+1):
        if a%x == 0:
            c=c+1
    if c == 2:
        return print('Prime')
    else:
        return print('Not Prime')

print(prime(13))
print(prime(10))
```

```python
#pyramid @ with depth n 
def py_c(depth):
    c=0
    for x in range (depth,-1,-1):
        print(x*" "+"*"*(2*c-1)+x*" ") 
        c=c+1

py_c(5)
#-> upto certain depth - pattern is obv
#-> we need uniform space 3,2,1,0 in successive depths for depth = 4 
#->stars need are 2N-1 for step 
#-> combination of how spaces and stars are progressing 
```

```python
#trailing zeros in n factorial n! 
def trail(n):
    fac = 1
    for x in range(1,n+1):
        fac = fac*x
    count = 0
    a = fac 
    while True:
        r = fac%10
        if r == 0:
            count = count+1
        else:
            break 
       # fac=int(fac/10) #floating point division and type casting creates issue 
        fac=fac//10
    return a,count 
print(trail(20)) 
print(trail(25)) 
```
